```python
#Rearrange the columns in the dataframe as needed
#Rearrange columns by ordering them as expected and saving in a new variable
df1_arranged = df1[["CUSTOMER", "ST" ,"GENDER" , "EDUCATION","CUSTOMER_LIFETIME_VALUE" , "INCOME" , "NUMBER_OF_OPEN_COMPLAINTS" , "MONTHLY_PREMIUM_AUTO" , "POLICY_TYPE" , "VEHICLE_CLASS" , "TOTAL_CLAIM_AMOUNT"]]
df1_arranged
```

```python
# For df3 we firstly need to rename the 'STATE' column to 'ST':
df3 = df3.rename(columns={'STATE':'ST'})

df3
```

```python
df3_arranged = df3[["CUSTOMER", "ST" , "GENDER" , "EDUCATION" , "CUSTOMER_LIFETIME_VALUE" , "INCOME" , "NUMBER_OF_OPEN_COMPLAINTS" , "MONTHLY_PREMIUM_AUTO" , "POLICY_TYPE" , "VEHICLE_CLASS" , "TOTAL_CLAIM_AMOUNT"]]
df3_arranged
```

```python
#Concatenate the three dataframes
df = pd.concat([df1_arranged, df2_arranged, df3_arranged], axis=0)
df
```

```python
# Which columns are numerical?
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_columns)
```

```python
# Which columns are categorical?
categorical_columns = df.select_dtypes(include='object').columns
print(categorical_columns)
```

```python
#Delete the column education and the number of open complaints from the dataframe.
del df['EDUCATION']
del df['NUMBER_OF_OPEN_COMPLAINTS']
```

```python
display(df)
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

# Remove«ing the % character
df['CUSTOMER_LIFETIME_VALUE'] = df['CUSTOMER_LIFETIME_VALUE'].str.replace('%', '')

#Convert the values in the column to numeric data type using the 'pd.to_numeric()' function. Set the errors parameter to 'coerce' to handle any non-numeric values. This will convert non-numeric values to NaN.
df['CUSTOMER_LIFETIME_VALUE'] = pd.to_numeric(df['CUSTOMER_LIFETIME_VALUE'], errors='coerce')
```

```python
#Multiply the values in the column by 100 to adjust for the percentage conversion
df['CUSTOMER_LIFETIME_VALUE'] = df['CUSTOMER_LIFETIME_VALUE'] * 100

df
```

```python
#Check for duplicate rows in the data and remove if any.
duplicates = df.duplicated()
count_duplicates = duplicates.sum()
print(count_duplicates)
```

```python
#Removing the duplicates
df = df.drop_duplicates()
display(df)
```

```python
#Filter out the data for customers who have an income of 0 or less.
filtered_df = df[df['Income'] <= 0]
```
